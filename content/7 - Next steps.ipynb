# Different ways to run Jupyter Lab

## In your browser (JupyterLite)
In this setup, the Python kernel and the Jupyter Lab interface run completely __inside__ your browser. File storage also happens within the browser. There is no separate application running on your computer, or on a server.

The demo that you're currently viewing runs this way: https://faulkner-dan.github.io/jupyterlite-demo/lab/index.html

There are similar setups here: https://jupyter.org/try-jupyter/lab/

and here: https://jupyterlite.readthedocs.io/en/latest/_static/lab/

__Pros__
- Quick and easy
- Nothing to install
- Confidentiality: All information is saved locally (not in the cloud)
- Good for simple demos

__Cons__
- Implementation is experimental, so it may have more bugs
- No easy access to files on your local computer
- No simple way to share files with others
- Sometimes requires small code tweaks, so it isn't 100% compatible with other Jupyter setups


## Installed locally on your computer (Anaconda)
Anaconda is a widely-used Python distribution: https://www.anaconda.com/download<br>

Installation is straightforward. This will give you a complete, free-standing Python setup including Jupyter Lab. It will use your browser for the user interface, but under the hood there's a Python kernel running elsewhere on your computer.

__Pros__
- A full Python/Jupyter setup that is completely under your control
- Confidentiality: All information is saved locally (not in the cloud)
- Lots of help available online
- Can integrate with version control systems (Git, GitHub, Azure DevOps, etc.)
- No network connection required

__Cons__
- Requires installation
- Limited built-in abilities to share and collaborate

- A full Python/Jupyter setup with lots of help available online
- Confidentiality: All information is saved on company-controlled servers (not in the cloud)

__Cons__
- Requires IT work to setup and maintain


## Cloud-based
<div class="alert alert-block alert-danger">
<b>Be careful:</b> Do not use cloud-based systems for work involving confidential information
</div>

There are several companies that offer this (I haven't tried all of these):

[Google Colab](https://colab.research.google.com/)

[Anaconda 'Code in the Cloud'](https://www.anaconda.com/code-in-the-cloud)

[Kaggle](https://www.kaggle.com/)

[GitHub Codespaces](https://github.com/codespaces)

...and probably many more.

__Pros__
- Nothing to install or maintain
- Most systems include built-in version control
- Easy to share and collaborate with other

__Cons__
- Confidentiality concerns
- Limitations on calculation speed and time. (When I tested Google Colab it seemed slow, especially when using interactive widgets.)
